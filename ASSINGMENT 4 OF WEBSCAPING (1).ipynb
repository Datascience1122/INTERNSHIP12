# QUESTION NO 7

```python
driver=webdriver.Chrome()
driver.get('https://www.imdb.com/search/title/?title_type=tv_series&sort=num_votes,desc ')
```

```python
Name=[]
Year_span=[]
Genre=[]
Rating=[]
Votes=[]
```

```python
name=driver.find_elements(By.XPATH,'//a[@class="ipc-title-link-wrapper"]')
for i in name:
    Name.append(i.text)
    
year=driver.find_elements(By.XPATH,'//div[@class="sc-be6f1408-7 iUtHEN dli-title-metadata"]/span[1]')  
for i in year:
    Year_span.append(i.text)
    
gen=driver.find_elements(By.XPATH,'//div[@class="ipc-html-content-inner-div"]')
for i in gen:
    Genre.append(i.text)
    
rating=driver.find_elements(By.XPATH,'//span[@class="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating"]')
for i in rating:
    Rating.append(i.text)
    
vote=driver.find_elements(By.XPATH,'//span[@class="ipc-rating-star--voteCount"]')
for i in vote:
    Votes.append(i.text)
```

```python
print(len(Name),len(Year_span),len(Genre),len(Rating),len(Votes))
```

```python
TV_Series=pd.DataFrame({'series name':Name,'genre':Genre,'rating':Rating,'votes':Votes})
TV_Series
```
